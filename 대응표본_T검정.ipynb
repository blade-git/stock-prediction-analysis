```python
import pandas as pd
from scipy import stats
```

```python
# 2개의 CSV 파일 경로
kospi_model_file = '/content/drive/MyDrive/주가데이터 거시경제변수 코스피 감성분석 조합 모델.csv'
macro_model_file = '/content/drive/MyDrive/주가데이터 거시경제변수 모델.csv'

# 각 파일에서 데이터를 불러오기
kospi_model_data = pd.read_csv(kospi_model_file, encoding='cp949')
macro_model_data = pd.read_csv(macro_model_file, encoding='cp949')
```

```python
kospi_model_data
```

```python
macro_model_data
```

```python
# 업종별로 데이터 프레임을 그룹화
kospi_grouped = kospi_model_data.groupby('업종')
macro_grouped = macro_model_data.groupby('업종')
```

```python
# 업종 리스트 확인 (두 데이터셋의 업종이 동일하다고 가정)
sectors = kospi_model_data['업종'].unique()
```

```python
# 업종별 대응표본 T검정 수행
for sector in sectors:
    kospi_sector_rmse = kospi_grouped.get_group(sector)['RMSE']  # 코스피 모델의 해당 업종 RMSE 값
    macro_sector_rmse = macro_grouped.get_group(sector)['RMSE']  # 거시경제 모델의 해당 업종 RMSE 값
    kospi_sector_mae = kospi_grouped.get_group(sector)['MAE']
    macro_sector_mae = macro_grouped.get_group(sector)['MAE']
    kospi_sector_mse = kospi_grouped.get_group(sector)['MSE']
    macro_sector_mse = macro_grouped.get_group(sector)['MSE']

    # RMSE에 대한 대응표본 T검정
    t_stat_rmse, p_value_rmse = stats.ttest_rel(kospi_sector_rmse, macro_sector_rmse)
    # MAE에 대한 대응표본 T검정
    t_stat_mae, p_value_mae = stats.ttest_rel(kospi_sector_mae, macro_sector_mae)
    # MSE에 대한 대응표본 T검정
    t_stat_mse, p_value_mse = stats.ttest_rel(kospi_sector_mse, macro_sector_mse)

    # 결과 출력
    print(f"업종: {sector}")


    # RMSE 결과
    print(f"RMSE - T-Statistic: {t_stat_rmse}, P-Value: {p_value_rmse}")
    if p_value_rmse < 0.1:
        print(f"{sector} 업종에서 두 모델 간의 RMSE 평균 차이가 통계적으로 유의미합니다.")
    else:
        print(f"{sector} 업종에서 두 모델 간의 RMSE 평균 차이가 통계적으로 유의미하지 않습니다.")

    print("=" * 50)

    # MAE 결과
    print(f"MAE - T-Statistic: {t_stat_mae}, P-Value: {p_value_mae}")
    if p_value_mae < 0.1:
        print(f"{sector} 업종에서 두 모델 간의 MAE 평균 차이가 통계적으로 유의미합니다.")
    else:
        print(f"{sector} 업종에서 두 모델 간의 MAE 평균 차이가 통계적으로 유의미하지 않습니다.")

    print("=" * 50)

    # MSE 결과
    print(f"MSE - T-Statistic: {t_stat_mse}, P-Value: {p_value_mse}")
    if p_value_mse < 0.1:
        print(f"{sector} 업종에서 두 모델 간의 MSE 평균 차이가 통계적으로 유의미합니다.")

    else:
        print(f"{sector} 업종에서 두 모델 간의 MSE 평균 차이가 통계적으로 유의미하지 않습니다.")

    print("=" * 50)
```
